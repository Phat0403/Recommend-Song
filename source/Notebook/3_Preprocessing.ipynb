#### Xử lý dữ liệu ngoại lai

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
numeric_columns = original.select_dtypes(include=['float64', 'int64']).columns
for column in numeric_columns:
    initial_rows = len(original)
    processed_data = remove_outliers_iqr(original, column)
    print(f"{column}: Đã loại bỏ {initial_rows - len(processed_data)} giá trị ngoại lai")
```
